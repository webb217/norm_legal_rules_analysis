```python
import pandas as pd
df = pd.read_csv('legal_rules_dataset_pk.csv')
```

```python
df_cat = pd.read_csv('rule_id_to_category.csv')
```

```python
df_full = df.merge(df_cat,on="rule_id",how="left")
```

```python
print(df_full.columns)
```

```python
cols = df_full.columns.tolist()
cols.remove("category")
insert_at = cols.index("rule_id")+1
cols.insert(insert_at,'category')

df_final = df_full[cols]
```

```python
df_final.to_csv('legal_rules_dataset_with_cat.csv',index=False)
```
